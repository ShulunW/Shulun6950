```python

from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

#the independent variables set with a column of constant 1. 
#Since it is required for running regression.

#The entire explantory variables with a constant column
X_vif = pd.DataFrame(add_constant(dfwinefinal.iloc[:,0:11]))

# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X_vif.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i)
                          for i in range(len(X_vif.columns))]
  
print(vif_data)


#Extra variables are removed based on the VIF criterion, here the removed variables are 
#residual.sugar, density,total sulfur dioxide and alcohol. 
#The adjusted data set has much better VIF information.

cols = [0,1,2,3,5,6,9,10]
X_vif_1= pd.DataFrame(add_constant(X_vif[X_vif.columns[cols]]))

# VIF dataframe
vif_data_1 = pd.DataFrame()
vif_data_1["feature adjusted"] = X_vif_1.columns
  
# calculating VIF for each feature
vif_data_1["VIF"] = [variance_inflation_factor(X_vif_1.values, i)
                          for i in range(len(X_vif_1.columns))]
  
print(vif_data_1)
```

```python
#Now we have the final dataset ready for building a logistic regression.
cols_1 = [0,1,2,3,5,6,9,10,11]
wwfinal=pd.DataFrame(add_constant(dfwinefinal[dfwinefinal.columns[cols_1]]))
wwfinal.head(10)

wwfinal['level'] = np.where(wwfinal['quality'] > 5, 'Fine', 'Poor') 

wwfinal.head(20)

wwfinal.to_hdf('whitewineFinal.h5', key='df', mode='w') #Saving the processed data into hd5 file.
```

```python
from sklearn.model_selection import train_test_split
#Ignoring quality variable
X_lo = wwfinal.iloc[:,0:8]

#Treat the responde variable as categorical variable
y_lo = wwfinal['level'].astype('category')

#Spliting the final dataframe in to testing and training with its corresponding x and y 
X_train,X_test,y_train,y_test = train_test_split(X_lo,y_lo,test_size=0.2,random_state=92815)
```

```python
from sklearn.linear_model import LogisticRegression


fit_lo= LogisticRegression(solver='lbfgs', max_iter=1000)
#A model is now built using 80% of the final dataset, which we called training set
fit_lo.fit(X_train,y_train)

#Our predicted category based on the logistics model using the training dataset.
y_pred=fit_lo.predict(X_test)

#Now we have the predicted value using the training model, an also the actual value
#we ignored from the testing set, now the goal is to build a condusion matrix.
```
