```python
#Outliers detection using cooks distance

#The formula for Cookâ€™s distance is:

#Di = (ri2 / p*MSE) * (hii / (1-hii)2)

#where:

#ri is the ith residual
#p is the number of coefficients in the regression model
#MSE is the mean squared error
#hii is the ith leverage value

import numpy as np
np.set_printoptions(suppress=True)

#create instance of influence
influence = fit_linear.get_influence()

#obtain Cook's distance for each observation
cooks = influence.cooks_distance

#display Cook's distances
print(cooks)

#Plotting cooks distance
'''
plt.figure(figsize=(10, 7))
plt.plot(list(range(0,4898)), cooks[0],"o")
plt.xlim([0, 5000])
plt.xlabel('Wine ID')
plt.ylabel('Cooks Distance value')
plt.show()


class cook_dis:
    def __init__(self,range0,range1,range2,range3,range4,x_label2,y_label2):
        self.range0=range0
        self.range1=range1
        self.range2=range2
        self.range3=range3
        self.range4=range4
        self.x_label2=x_label2
        self.y_label2=y_label2
        
    def scatter(self):
        plt.figure(figsize=(self.range0,self.range1))
        plt.plot(self.range2, self.range3,"o")
        plt.xlim(self.range4)
        plt.xlabel(self.x_label2)
        plt.ylabel(self.y_label2)
        plt.show()
 '''
from classes import cook_dis
distance1 = cook_dis(10, 7,list(range(0,4898)),cooks[0],[0, 5000],'Wine ID','Cooks Distance value')
distance1.scatter()

#Judging by cooks distance plot there exist two obvious outliers, so the next step is to remove them.
```

```python
#Before building the model, it is important to check if there exist multicolinearality. We obtain the VIF information for this dataframe.

#Multicollinearity occurs when there are two or more independent variables in a
#multiple regression model, which have a high correlation among themselves. 
#When some features are highly correlated, we might have difficulty in distinguishing 
#between their individual effects on the dependent variable. Multicollinearity can be 
#detected using various techniques, one such technique being the Variance Inflation Factor(VIF).

#VIF is calculated by 1/(1-R_squared)

from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# the independent variables set with a column of constant 1. Since it is required for running regression.
X_vif = pd.DataFrame(add_constant(dfwinefinal.iloc[:,0:11]))

# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X_vif.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i)
                          for i in range(len(X_vif.columns))]
  
print(vif_data)


#Extra variables are removed based on the VIF criterion, here the removed variables are 
#residual.sugar, density,total sulfur dioxide and alcohol. 
#The adjusted data set has much better VIF information.

cols = [0,1,2,3,5,6,9,10]
X_vif_1= pd.DataFrame(add_constant(X_vif[X_vif.columns[cols]]))

# VIF dataframe
vif_data_1 = pd.DataFrame()
vif_data_1["feature adjusted"] = X_vif_1.columns
  
# calculating VIF for each feature
vif_data_1["VIF"] = [variance_inflation_factor(X_vif_1.values, i)
                          for i in range(len(X_vif_1.columns))]
  
print(vif_data_1)
```

```python
#Now we have the final dataset ready for building a logistic regression.
cols_1 = [0,1,2,3,5,6,9,10,11]
wwfinal=pd.DataFrame(add_constant(dfwinefinal[dfwinefinal.columns[cols_1]]))
wwfinal.head(10)

#Since we are missing lots of observations from quality level 8, 4, 3, and 9, it is not accurate to treat 
#all levels as categorical variables. As a result we devide the quality level into 2 categories, quality 
#larger than 5 will be categorized as Fine, less than 5 will be categorized as Poor.

wwfinal['level'] = np.where(wwfinal['quality'] > 5, 'Fine', 'Poor') 

wwfinal.head(20)
```
